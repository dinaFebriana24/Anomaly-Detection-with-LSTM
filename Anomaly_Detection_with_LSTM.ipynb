# Data Preprocessing

```python
import os
import pandas as pd
import time
from sklearn.preprocessing import MinMaxScaler
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path ke dataset
folder_path = "/content/drive/MyDrive/Dataset/Data"

# Baca dataset
data_path = os.path.join(folder_path, "data.csv")

# Cek apakah file ada sebelum membaca

if not os.path.exists(data_path):
     raise FileNotFoundError("File data.csv tidak ditemukan!")

# train = pd.read_csv(train_path)
data = pd.read_csv(data_path)

def convert_timestamp(ts):
    """Fungsi untuk mengonversi timestamp ke format standar YYYY-MM-DD HH:MM:SS.SSS"""
    if pd.isna(ts) or not isinstance(ts, str):
        return None  # Jika kosong atau bukan string, return None

    try:
        ts = ts.replace(" - ", "-")  # Perbaiki format tanggal jika ada spasi
        date_part, time_part = ts.split("_")  # Pisahkan tanggal dan waktu
        formatted_time = f"{time_part[:2]}:{time_part[2:4]}:{time_part[4:6]}.{time_part[7:]}"  # Format waktu
        return pd.to_datetime(f"{date_part} {formatted_time}", errors="coerce")  # Konversi ke datetime
    except Exception as e:
        print(f"Error processing timestamp: {ts} - {e}")
        return None

def preprocess_data(df, df_path):
    if df.empty:
        return df  # Jika dataframe kosong, langsung return

    # Cek apakah ada kolom timestamp, jika ada maka konversi
    timestamp_column = "Timestamp"  # Sesuaikan nama kolom timestamp di dataset
    if timestamp_column in df.columns:
        df[timestamp_column] = df[timestamp_column].apply(convert_timestamp)

    # Mapping label anomaly ke angka
    if 'Anomaly' in df.columns:
        df['Anomaly'] = df['Anomaly'].replace({'-': 0, 'Freeze': 1, 'Ramp': 2, 'Step': 3})

    # Simpan target variabel
    Y = df['Anomaly'] if 'Anomaly' in df.columns else None

    # Hapus kolom yang tidak digunakan
    drop_columns = ['number', 'Parameter for Anomaly', 'Actual value', 'Run id']
    df = df.drop(columns=[col for col in drop_columns if col in df.columns], errors='ignore')

    # Normalisasi hanya kolom numerik
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    scaler = MinMaxScaler()

    if numeric_cols:
        df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

    # Simpan scaler hanya jika sedang memproses data
    if 'data' in df_path:
        scaler_path = "/content/drive/MyDrive/Dataset/PreprocessingData/scaler.pkl"
        pd.to_pickle(scaler, scaler_path)
        print("Scaler berhasil disimpan di", scaler_path)

    # Tambahkan kembali timestamp jika ada
    if timestamp_column in df.columns:
        df[timestamp_column] = df[timestamp_column]

    # Kembalikan label 'Anomaly' ke dataset jika ada
    if Y is not None:
        df['Anomaly'] = Y

    return df

# Mulai pencatatan waktu pemrosesan
start = time.time()

# Preprocess data
data_processed = preprocess_data(data, data_path)

# Buat folder untuk penyimpanan jika belum ada
output_folder = "/content/drive/MyDrive/Dataset/PreprocessingData"
os.makedirs(output_folder, exist_ok=True)

# Simpan data yang sudah diproses ke pickle
data_pickle_path = os.path.join(output_folder, "data.pkl")

data_processed.to_pickle(data_pickle_path)

# Cek apakah file berhasil disimpan
if os.path.exists(data_pickle_path):
    print("File data.pkl berhasil disimpan!")
else:
    print("File data.pkl gagal disimpan!")

# Hitung waktu eksekusi
end = time.time()
print("Total time to process and save data: {:.2f} seconds".format(end - start))
```

```python
import pandas as pd

# Load dataset
data = pd.read_pickle("/content/drive/MyDrive/Dataset/PreprocessingData/data.pkl")

# Konversi timestamp agar pasti dalam format datetime
time_column = "Timestamp"  # Sesuaikan dengan nama kolom timestamp
data[time_column] = pd.to_datetime(data[time_column], errors='coerce')

# **Sort data berdasarkan timestamp sebelum split**
data = data.sort_values(by=time_column).reset_index(drop=True)

# Pisahkan data berdasarkan kategori Anomaly
train_splits = []
val_splits = []

for category in data['Anomaly'].unique():
    category_data = data[data['Anomaly'] == category]  # Ambil data per kategori
    split_idx = int(len(category_data) * 6 / 7)  # Tentukan index split 6:1

    train_splits.append(category_data.iloc[:split_idx])  # Train data per kategori
    val_splits.append(category_data.iloc[split_idx:])    # Validation data per kategori

# Gabungkan kembali data train & validation dari semua kategori
train_data = pd.concat(train_splits).sort_values(by=time_column).reset_index(drop=True)
val_data = pd.concat(val_splits).sort_values(by=time_column).reset_index(drop=True)

# Simpan hasil split dalam format PKL
train_data.to_pickle("/content/drive/MyDrive/Dataset/PreprocessingData/train.pkl")
val_data.to_pickle("/content/drive/MyDrive/Dataset/PreprocessingData/val.pkl")

# **Cek kembali apakah data sudah terurut**
print("✅ Data telah disortir berdasarkan waktu sebelum split!")

# Cek apakah data train dan validation masih dalam urutan yang benar
print(f"Apakah Train terurut waktu? {'✅ Ya' if train_data[time_column].is_monotonic_increasing else '❌ Tidak'}")
print(f"Apakah Validation terurut waktu? {'✅ Ya' if val_data[time_column].is_monotonic_increasing else '❌ Tidak'}")

# Cek rentang waktu setelah sorting
print("\n🔍 Rentang waktu setelah sorting:")
print(f"Train: {train_data[time_column].min()} - {train_data[time_column].max()}")
print(f"Val  : {val_data[time_column].min()} - {val_data[time_column].max()}")

# Cek jumlah kategori Anomaly di masing-masing dataset
print("\nJumlah Kategori Anomaly di Train Data:")
print(train_data['Anomaly'].value_counts())

print("\nJumlah Kategori Anomaly di Validation Data:")
print(val_data['Anomaly'].value_counts())
```

# BiLSTM Anomaly Detection

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Bidirectional, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from google.colab import drive

# **Mount Google Drive**
drive.mount('/content/drive')

# **Path ke dataset**
train_path = "/content/drive/MyDrive/Dataset/PreprocessingData/train.pkl"
val_path = "/content/drive/MyDrive/Dataset/PreprocessingData/val.pkl"
test_path = "/content/drive/MyDrive/Dataset/PreprocessingData/test_sorted.pkl"

# **Load dataset**
df_train = pd.read_pickle(train_path)
df_val = pd.read_pickle(val_path)
df_test = pd.read_pickle(test_path)

# **Pastikan data tidak kosong**
assert not df_train.empty, "Train dataset kosong!"
assert not df_val.empty, "Validation dataset kosong!"
assert not df_test.empty, "Test dataset kosong!"

# **Mengubah label anomaly menjadi binary classification**
df_train['Anomaly'] = df_train['Anomaly'].apply(lambda x: 1 if x in [1, 2, 3] else 0)
df_val['Anomaly'] = df_val['Anomaly'].apply(lambda x: 1 if x in [1, 2, 3] else 0)
df_test['Anomaly'] = df_test['Anomaly'].apply(lambda x: 1 if x in [1, 2, 3] else 0)

# **Memisahkan fitur (X) dan target (y)**
X_train = df_train.drop(['Anomaly', 'Timestamp'], axis=1)
y_train = df_train['Anomaly'].values
X_val = df_val.drop(['Anomaly', 'Timestamp'], axis=1)
y_val = df_val['Anomaly'].values
X_test = df_test.drop(['Anomaly', 'Timestamp'], axis=1)
y_test = df_test['Anomaly'].values

# **Mengatasi error dtype object dengan konversi ke numerik**
X_train = X_train.apply(pd.to_numeric, errors='coerce').fillna(0)
X_val = X_val.apply(pd.to_numeric, errors='coerce').fillna(0)
X_test = X_test.apply(pd.to_numeric, errors='coerce').fillna(0)

# **Membentuk sequence data dengan time-steps = 5**
def create_time_window(X, y, time_steps=5):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X.iloc[i: i + time_steps].values)
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

X_train, y_train = create_time_window(X_train, y_train, time_steps=5)
X_val, y_val = create_time_window(X_val, y_val, time_steps=5)
X_test, y_test = create_time_window(X_test, y_test)

# **Model BiLSTM**
model = Sequential([
    Input(shape=(5, X_train.shape[2])),
    Bidirectional(LSTM(64, return_sequences=True)),
    BatchNormalization(),
    Dropout(0.3),

    Bidirectional(LSTM(32, return_sequences=True)),
    BatchNormalization(),
    Dropout(0.3),

    Bidirectional(LSTM(16, return_sequences=False)),
    BatchNormalization(),
    Dropout(0.3),

    Dense(32, activation='relu'),
    Dropout(0.2),

    Dense(16, activation='relu'),
    Dropout(0.2),

    Dense(1, activation='sigmoid')
])

# **Compile model**
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# **Callback Early Stopping**
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# **Training model**
history = model.fit(X_train, y_train, epochs=50, batch_size=128, validation_data=(X_val, y_val),
                    verbose=1, callbacks=[early_stopping])

# **Evaluasi model pada test set**
y_test_pred = (model.predict(X_test) > 0.5).astype(int).flatten()

y_test = y_test.astype(int)  # Convert y_test to int if it's not already
y_test_pred = y_test_pred.astype(int)  # Convert y_test_pred to int

# **Menghitung metrik evaluasi untuk test set**
test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred)
test_recall = recall_score(y_test, y_test_pred)
test_f1 = f1_score(y_test, y_test_pred)
test_conf_matrix = confusion_matrix(y_test, y_test_pred)

# **Menampilkan hasil evaluasi test set**
print(f"✅ Test Accuracy: {test_accuracy:.4f}")
print(f"✅ Test Precision: {test_precision:.4f}")
print(f"✅ Test Recall: {test_recall:.4f}")
print(f"✅ Test F1-Score: {test_f1:.4f}")
print("\nConfusion Matrix (Test Set):")
print(test_conf_matrix)
print("\nClassification Report (Test Set):")
print(classification_report(y_test, y_test_pred))

# **Visualisasi Confusion Matrix untuk test set**
plt.figure(figsize=(6, 4))
sns.heatmap(test_conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Normal", "Anomaly"], yticklabels=["Normal", "Anomaly"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix (Test Set)")
plt.show()

# **Simpan model**
model.save("/content/drive/MyDrive/Dataset/PreprocessingData/lstm_anomaly_model_fixed.h5")
print("✅ Model saved successfully!")
```

# BiLSTM + Cross Validation

```python
# === IMPORT LIBRARY ===
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Bidirectional, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from google.colab import drive

# === MOUNT GOOGLE DRIVE ===
drive.mount('/content/drive')

# === LOAD DATA ===
train_path = "/content/drive/MyDrive/Dataset/PreprocessingData/train.pkl"
val_path = "/content/drive/MyDrive/Dataset/PreprocessingData/val.pkl"
test_path = "/content/drive/MyDrive/Dataset/PreprocessingData/test_sorted.pkl"

df_train = pd.read_pickle(train_path)
df_val = pd.read_pickle(val_path)
df_test = pd.read_pickle(test_path)

assert not df_train.empty, "Train dataset kosong!"
assert not df_val.empty, "Validation dataset kosong!"
assert not df_test.empty, "Test dataset kosong!"

# === PREPROCESSING ===
for df in [df_train, df_val, df_test]:
    df['Anomaly'] = df['Anomaly'].apply(lambda x: 1 if x in [1, 2, 3] else 0)

def prepare_xy(df):
    X = df.drop(['Anomaly', 'Timestamp'], axis=1)
    X = X.apply(pd.to_numeric, errors='coerce').fillna(0)
    y = df['Anomaly'].values
    return X, y

# Gabungkan train + val agar TimeSeriesSplit bisa digunakan penuh
df_all = pd.concat([df_train, df_val], ignore_index=True)
X_all, y_all = prepare_xy(df_all)

# === CREATE SEQUENCE WINDOWS ===
def create_time_window(X, y, time_steps=5):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X.iloc[i: i + time_steps].values)
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

time_steps = 5
X_all_seq, y_all_seq = create_time_window(X_all, y_all, time_steps)

# === TIME SERIES SPLIT CROSS-VALIDATION ===
tscv = TimeSeriesSplit(n_splits=7)
fold_metrics = []

for fold, (train_idx, val_idx) in enumerate(tscv.split(X_all_seq)):
    print(f"\n🔄 Fold {fold + 1}")

    X_fold_train, X_fold_val = X_all_seq[train_idx], X_all_seq[val_idx]
    y_fold_train, y_fold_val = y_all_seq[train_idx], y_all_seq[val_idx]

    model = Sequential([
        Input(shape=(time_steps, X_fold_train.shape[2])),
        Bidirectional(LSTM(64, return_sequences=True)),
        BatchNormalization(),
        Dropout(0.3),

        Bidirectional(LSTM(32, return_sequences=True)),
        BatchNormalization(),
        Dropout(0.3),

        Bidirectional(LSTM(16, return_sequences=False)),
        BatchNormalization(),
        Dropout(0.3),

        Dense(32, activation='relu'),
        Dropout(0.2),

        Dense(16, activation='relu'),
        Dropout(0.2),

        Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

    model.fit(X_fold_train, y_fold_train,
              epochs=10,
              batch_size=128,
              validation_data=(X_fold_val, y_fold_val),
              callbacks=[early_stopping],
              verbose=1)

    y_val_pred = (model.predict(X_fold_val) > 0.5).astype(int).flatten()

    fold_result = {
        'accuracy': accuracy_score(y_fold_val, y_val_pred),
        'precision': precision_score(y_fold_val, y_val_pred, zero_division=0),
        'recall': recall_score(y_fold_val, y_val_pred, zero_division=0),
        'f1': f1_score(y_fold_val, y_val_pred, zero_division=0)
    }
    fold_metrics.append(fold_result)
    print(f"✅ Fold {fold + 1} Metrics:", fold_result)

# === AVERAGE CV METRICS ===
avg_metrics = pd.DataFrame(fold_metrics).mean().to_dict()
print("\n📊 Average Cross-Validation Metrics:")
for metric, value in avg_metrics.items():
    print(f"{metric.capitalize()}: {value:.4f}")


# === SIMPAN MODEL TERAKHIR ===
model.save("/content/drive/MyDrive/Dataset/PreprocessingData/lstm_anomaly_model_tscv_final.h5")
print("✅ Model saved successfully!")
```
